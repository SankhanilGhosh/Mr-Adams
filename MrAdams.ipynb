### Title
- Rule Based AI Chatbot using python
### Objective
- To create a chatbot that can respond to basic user queries
### Concept
- Uses if else command (rule based AI)
- Matches user input with preferred response

```python
def MrAdmas():
    print("MrAdmas: Hello! Type 'bye' to exit.")

    while True:
        user = input("You: ").lower().strip()

        if not user:
            continue

        if "hello" in user:
            print("MrAdmas: Hi! How can I help you?")
        
        elif "how are you" in user:
            print("MrAdmas: I am fine!")
        
        elif "what is ai" in user:
            print("MrAdmas: AI means Artificial Intelligence.")
        
        elif "bye" in user:
            print("MrAdmas: Goodbye!")
            break
        
        else:
            print("MrAdmas: Sorry, I don't understand.")

MrAdmas()
```
